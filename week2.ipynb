```python
def my_f(n=100): #n sayısını arttırdıkça 0.5'e yaklaşıyor.
    h_number=0
    t=[]
    for i in range(n):
        outcome=np.random.choice(['H','T'])
        t.append(outcome)
        #print(i,"th experiment:",o)
        if outcome=='H':
            h_number+=1
    collections.Counter(t)
    return h_number/n
```
